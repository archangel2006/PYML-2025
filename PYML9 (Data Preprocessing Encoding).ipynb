# ORDINAL ENCODING

```python
#!pip install scikit-learn

# Import Related CLass & Module

from sklearn.preprocessing import OrdinalEncoder
```

```python
# CREATE DATAFRAME

data={'Student':['Alice','Bob','Charlie','David','Eva'],
      'Grade':['A','B','C','A','B']}
df1=pd.DataFrame(data)
print(df1)
```

```python
# INITIALIZE & APPLY OrdinalEncoder

# Creates an Ordinal Encoder object from scikit-learn
# Tells it to treat 'A'<'B'<'C' ,i.e, defining a CUSTOM ORDER.
# Encode 'A'→0,'B'→1,'C'→2
# encoder.fit_transform(...) : Learns the mapping between categorical labels (A,B,C) and applies the transformation.

encoder=OrdinalEncoder(categories=[['A','B','C']])
df1['GradeEncoded']==encoder.fit_transform(df1[['Grade']])
print(df1)
```

# LABEL ENCODING



*   Label encoding is ideal when the categories are nominal, meaning there is no rank or hierarchy among them



```python
# Import Related CLass & Module

from sklearn.preprocessing import LabelEncoder

# LabelEncoder is a class
# sklearn.preprocessing is a module
```

```python
df2 = pd.DataFrame({ 'CustomerID': [101, 102, 103, 104, 105, 106, 107],
                     'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
                      'City': ['New York', 'Paris', 'Berlin', 'New York', 'Paris', 'Berlin', 'Paris'],
                     'Purchased': ['Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No'] })

print(df2)
```

```python
# Apply LabelEncoder to 'Gender' and 'City'

le_Gender=LabelEncoder()
le_City=LabelEncoder()

# Apply LabelEncoder in-place ,i.e, overwrite the existing 'Purchased' Column instead of creating a new one.
le_Purchased=LabelEncoder()

## le_Gender:A variable that stores the new instance of LabelEncoder. You can now use it to encode data.
```

```python
#encodes the 'Gender' column in your DataFrame (data) into numeric labels and saves the result in a new column called 'Gender_encoded'.

df2['Gender_Encoded']=le_Gender.fit_transform(df2['Gender'])
df2['City_Encoded']=le_City.fit_transform(df2['City'])
df2['Purchased']=le_Purchased.fit_transform(df2['Purchased'])

print(df2)
```

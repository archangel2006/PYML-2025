# Numerical Data

```python
fig=plt.figure(figsize=(6,4))
plt.hist(df['Age'],bins=40)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Split the age range into 50 equal-width intervals (bins)
```

```python
fig=plt.figure(figsize=(6,4))
plt.hist(df['Age'].dropna(),bins=40,color='lightblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Split the age range into 50 equal-width intervals (bins)
```

```python
fig=plt.figure(figsize=(4,4))
sns.displot(df['Age'].dropna(),bins=30,kde=True,color='coral')
plt.show()

# kde=True : Kernel Density display =True
```

```python
df['Age'].skew()
```

```python
fig=plt.figure(figsize=(4,4))
sns.displot(df['Fare'].dropna(),bins=30,kde=True,color='teal')
plt.show()
df['Fare'].skew()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
sns.boxplot(df['Age'],ax=axs[0])
sns.boxplot(df['Age'],orient='h',ax=axs[1])
plt.show()
```

```python
Q1=df['Age'].quantile(0.25)
Q2=df['Age'].quantile(0.50)
Q3=df['Age'].quantile(0.75)
#Q4=df['Age'],quantile(1)
print(f"Quatile1:{Q1}")
print(f"Quatile2:{Q2}")
print(f"Quatile3:{Q3}")

print(f"Median:{df['Age'].median()}")
print(f"Minimum: {df['Age'].min()}")
print(f"Maximum:{df['Age'].max()}")
print(f"Mean: {df['Age'].mean()}")
print(f"Skewness: {df['Age'].skew()}")
```

```python
sns.pairplot(iris,hue='species',diag_kind='kde')
plt.suptitle('Pair Plot Of Iris Dataset') # Super Title
plt.show()
```

```python
corr=iris.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')  # annot=True : we're displaying values
plt.title('Correlation Matrix')
plt.show()
```
